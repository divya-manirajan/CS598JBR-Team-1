```python
import pandas as pd
import json

base_result = "base_prompt_248264623217713294771470079546488913479.jsonl_results.jsonl" 
instruct_result = "instruct_prompt_248264623217713294771470079546488913479.jsonl_results.jsonl"
# Load JSON data from two files
base_df = pd.read_json(base_result, lines=True)
instruct_df = pd.read_json(instruct_result, lines=True)
df = pd.merge(base_df, instruct_df, on='task_id', suffixes=('_base', '_instruct'))[['task_id', 'passed_base', 'passed_instruct']]
df['num'] = df['task_id'].str.split('/').str[1].astype(int)
df.sort_values(by='num', inplace=True)
df.drop(columns='num', inplace=True)
df
```

```python
latex_table = df.to_latex(index=False)
print(latex_table)
```
