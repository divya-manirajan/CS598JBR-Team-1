```python
import jsonlines
import pandas as pd
seed = "248264623217713294771470079546488913479"
vanilla_log = "task_1_" + seed + "_vanilla.jsonl"
crafted_log = "task_1_" + seed + "_crafted.jsonl"

vanilla_df = pd.read_json(vanilla_log, lines=True)
crafted_df = pd.read_json(crafted_log, lines=True)
all_df = pd.merge(vanilla_df, crafted_df, on="task_id", suffixes=("_vanilla", "_crafted"))
all_df.info()
```

```python
accuracy_vanilla = all_df['is_correct_vanilla'].value_counts()[True] / all_df["is_correct_vanilla"].size
accuracy_crafted = all_df['is_correct_crafted'].value_counts()[True] / all_df["is_correct_crafted"].size
print("Vanilla accuracy: ", accuracy_vanilla)
print("Crafted accuracy: ", accuracy_crafted)
```

```python
df = all_df[['task_id','compiled_vanilla','ran_vanilla','is_correct_vanilla','compiled_crafted','ran_crafted','is_correct_crafted']]
df
```

```python
up = '$\\uparrow$'
down = '$\\downarrow$'
even = '$-$'
df.loc[:,'delta'] = pd.Series([up if (df['is_correct_vanilla'][i] < df['is_correct_crafted'][i]) \
                          else down if (df['is_correct_vanilla'][i] < df['is_correct_crafted'][i]) \
                          else even for i in range(len(df))])
df
```
